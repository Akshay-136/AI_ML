**Outlier Detection**

```python
# z score , percentile , iqr --> removal methods (don't go for this method)
# replacement method - median
```

```python
sns.boxplot(df.bmi) #the points outside the line are known as outliers
```

```python
#iqr
'''
iqr = q3-q1
q1=df.bmi.quantile(0.25)
q3= df.bmi.quantile(0.75)

upper_limit=q3+1.5*IQR
Lower limit = q1-1.5*IQR


'''
```

```python
#outlier removal using IQR
```

```python
q1=df.bmi.quantile(0.25)
q3=df.bmi.quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR = q3-q1
IQR
```

```python
upper_limit = q3+1.5*IQR #upper limit bcz all the points are outside upper limit
upper_limit
```

```python
df = df[df.bmi<upper_limit]
```

```python
sns.boxplot(df.bmi) # so now we can see all outliers are removed
```

```python
df.shape #as we can see the numbers reduced as outliers removed
```
