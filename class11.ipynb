```python
df = pd.read_csv('/content/Iris.csv')
df.head()
```

```python
#checking for null values
df.isnull().sum()
```

```python
df['Species'].unique() #finding unique category
```

```python
#count check
df['Species'].value_counts()
```

```python
#splitting the data
x=df.iloc[:,1:5]
x
```

```python
y=df.iloc[:,5:]
y
```

```python
y=df['Species']
y[0:5]
```

```python
y = pd.get_dummies(df.iloc[:,5:]).values
y
```

```python
#splitting the data into train and test
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2 , random_state=21)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
'''le = LabelEncoder()
ytrain = le.fit_transform(ytrain)
ytest = le.transform(ytest) '''
```

```python
#ANN Model
model = Sequential()
model.add(Dense(4,input_dim=4,activation='relu')) #cols are 4 therefore input dimension = 4
model.add(Dense(32,activation='relu'))
model.add(Dense(26,activation='relu'))
model.add(Dense(3,activation='softmax')) #multiclass thats why softmax func
```

```python
model.compile(optimizer='adam',loss ='sparse_categorical_crossentropy',metrics=['accuracy'])

#regression we use "mse" in loss and activation func to be linear and in classification loss = categorical_crossentropy nd activation = softmax
```

```python
model.summary()
```

```python
model.fit(xtrain,ytrain,epochs=10,batch_size=5,validation_data=(xtest,ytest))
```

11th **lecture**

```python
ypred = model.predict([[5.5,2.34,5.3,4.3]])
print(ypred)
```

```python
ypred = np.argmax(ypred)
output = ['Iris-setosa','Iris-versicolor','Iris-virginica']
output[ypred]
```
